##### Limpieza de texto y tokenización.

Para calcular la frecuencia de las palabras en el texto, primero debemos limpiar el texto, dividiendo el texto en palabras, eliminando signos de puntuacion y convirtiendo todas las palabras a minusculas para asegurar que cuente correctamente la frecuencia. para esto usamos la funcion word_tokenize, que nos limpia el texto de forma predeterminada. 

```python
# TOKENIZACIÓN DEL TEXTO EN PALABRAS. (divide el texto en palabras)
tokens = word_tokenize(BCN.lower())
print(tokens)
print(f'\nNombre de paraules al text: {len(tokens)}')
```

```python
tokens_filtrados = [word for word in tokens if word.isalpha()] # Eliminamos los signos de puntuación, para luego contabilizar.
print(tokens_filtrados)
```

```python
frecuencia_palabras = Counter(tokens_filtrados) #(contabiliza, cuantas veces se repite cada "palabra" de caracteres alfabeticos.)
print(frecuencia_palabras)
```

```python
top_15_palabras = frecuencia_palabras.most_common(15)
top_15_palabras
```

las 15 palabras con mas frecuencia 

```python
palabras = [tupla[0] for tupla in top_15_palabras]
frecuencias = [tupla[1] for tupla in top_15_palabras]

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
plt.barh(palabras, frecuencias, color='skyblue')
plt.xlabel('Frecuencia')
plt.ylabel('Palabra')
plt.title('Frecuencia de Palabras')
plt.gca().invert_yaxis()  # Invertir el eje y para que la palabra más frecuente esté en la parte superior
plt.savefig('frecuencia_palabras.png')
plt.show()
```

Podemos ver en este grafico que las palabras con mas frecuencia son las que sirven de enlaces entre palabras o llamadas stopword, pero estas palabras no me aportan informacion relevante para mi estudio. solo digital, barcelona y systems. me sirven para mi estudio.

##### Stopwords:
La eliminación de palabras irrelevantes es un paso crucial de preprocesamiento de texto en el análisis de sentimientos que implica eliminar palabras comunes e irrelevantes que probablemente no transmitan mucho sentimiento. Las palabras vacías son palabras que son muy comunes en un idioma y no tienen mucho significado, como "y", "el", "de" y "ello". Estas palabras pueden causar ruido y sesgar el análisis si no se eliminan.

```python
stop_words = set(stopwords.words('english'))
tokens_filtrados2 = [word for word in tokens_filtrados if word not in stop_words]
print(tokens_filtrados2)
```

Eliminamos las stopword, nos permite obtener solo las palabras relevantes de mi texto y que me dan informacion.

```python
frecuencia_palabras2 = Counter(tokens_filtrados2) #(contabiliza, cuantas veces se repite cada "palabra" de caracteres alfabeticos sin incluir las stopwords.)
print(frecuencia_palabras2)
```

##### Stemmer: 
Un stemmer es una herramienta utilizada en el procesamiento del lenguaje natural (NLP) que se utiliza para reducir las palabras a su raíz o forma base, conocida como "stem". Esto se hace eliminando los afijos de las palabras, como los sufijos y prefijos, para producir la forma base de una palabra.

El objetivo principal del stemming es reducir las palabras a su forma más básica, lo que ayuda a normalizar el texto y reducir la dimensionalidad en tareas de análisis de texto, como la recuperación de información, la clasificación de texto y el análisis de sentimientos.

```python
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in tokens_filtrados2]
print(stemmed_words)
```

```python
frecuencia_palabras3 = Counter(stemmed_words)
print(frecuencia_palabras3)
```

```python
top3_15_palabras = frecuencia_palabras3.most_common(15)
top3_15_palabras
```

```python
palabras = [tupla[0] for tupla in top3_15_palabras]
frecuencias = [tupla[1] for tupla in top3_15_palabras]

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
plt.barh(palabras, frecuencias, color='skyblue')
plt.xlabel('Frecuencia')
plt.ylabel('Palabra')
plt.title('Frecuencia de Palabras sin stop words y con stemmer')
plt.gca().invert_yaxis()  # Invertir el eje y para que la palabra más frecuente esté en la parte superior
plt.savefig('Frecuencia de Palabras sin stop words y con stemmer.png')
plt.show()
```
